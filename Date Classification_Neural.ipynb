Program to build a Deep Learning model from scratch. this involves a multiclassification of date fruits into different types 


```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder, minmax_scale
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_excel ('date_fruit.xlsx')

data. head ()
```

```python
print (data.shape)
print (data ['Class'].unique ())
```

```python
#splitting dataset into features and labels

x = data.drop ('Class', axis = 1)
y = data.loc [:, 'Class']
```

```python
#normalising X
x_scaled = minmax_scale (x)

#creating a dataframe for the new x

x = pd.DataFrame (x_scaled)

x.head ()
```

```python
#creating a label encoder object

encoder = LabelEncoder ()

y = encoder.fit_transform (y)
```

```python
#dividing data into training and testing 
x_train, x_temporary, y_train, y_temporary = train_test_split (x,y, train_size = 0.8)

x_val, x_test, y_val, y_test = train_test_split (x_temporary, y_temporary, train_size = 0.5)
```

```python
print (f' The Length of the Dataset is: {len (data)}')
print (f' The Length of the training Dataset is: {len (x_train)}')
print (f' The Length of the validation Dataset is: {len (x_val)}')
print (f' The Length of the test Dataset is: {len (x_test)}')
```
